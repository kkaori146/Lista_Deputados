### Pipeline de Deputado por Despesa total

```python
p1 = beam.Pipeline()

pipe1 = (
    p1
    |'Extrair do CSV'>> beam.io.ReadFromText('/content/lista_final.csv',skip_header_lines = 1 )
    |'Separador do csv' >> beam.Map(lambda record: record.split(','))
    |'Filtrar despesas'>> beam.Filter(lambda record: float(record[3])>0)
    |'Agregar as colunas'>> beam.Map(lambda record: (record[5],round(float(record[3]))))
    |'Construir uma nova tabela' >> beam.CombinePerKey(sum)
)

tabela = (
    {'Despesas totais':pipe1}
    |'Agrupar as pernas' >> beam.CoGroupByKey()
    |'Resultado' >>beam.Map(print)
  
)

p1.run()
```

### Modelo GCP

```python
# Despesas políticas entre 2002 - 2022 

import apache_beam as beam
import os 
from apache_beam.options.pipeline_options import PipelineOptions

pipeline_options ={
    'project':'euphoric-anchor-362800',
    'runner':'DataflowRunner',
    'region':'southamerica-east1',
    'staging_location':'gs://lista_deputados/staging/',
    'temp_location':'gs://lista_deputados/temp/',
    'template_location':'gs://lista_deputados/models/modelo_batch'
}

# Service Account GCP (Token)
serviceAccount = '/content/token.json'
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = serviceAccount

pipeline_options = PipelineOptions.from_dictionary(pipeline_options)

p1 = beam.Pipeline(options=pipeline_options)

despesas = (
    p1
    |'Extrair do CSV'>> beam.io.ReadFromText('gs://lista_deputados/lista_final.csv',skip_header_lines = 1)
    |'Separador do csv' >> beam.Map(lambda record: record.split(','))
    |'Filtrar despesas'>> beam.Filter(lambda record: float(record[3])>0)
    |'Agregar as colunas'>> beam.Map(lambda record: (record[0],round(float(record[3]))))
    |'Construir uma nova tabela' >> beam.CombinePerKey(sum)

)

tabela = (
    {'Despesas totais':despesas}
    |'Agrupar as pernas' >> beam.CoGroupByKey()
    |'Gravar o resultado'>> beam.io.WriteToText('gs://lista_deputados/tratados/lista',file_name_suffix='.csv')
)
p1.run()
```
